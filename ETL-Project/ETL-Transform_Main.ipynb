## Step I:  Merge

### Netflix
- Merge netflix_final and movies_revenue on title: **netflix_revenu_df**


### Budget
- Merge movies_budget_df and movies_budget_df_revenue on id: **movies_budget_revenue_df**

```python
netflix = pd.read_csv('netflix_final.csv')
netflix.head()
```

```python
movies_revenue = pd.read_csv('movie_revenue.csv')
movies_revenue.head()
```

```python
netflix_revenue_df = pd.merge(netflix, movies_revenue, on="title")
netflix_revenue_df.head()
```

```python
movie_num = pd.read_csv('movies_num_final.csv')
```

```python
movie_num = movie_num.rename(columns={"Movie":"title"})
movie_num.head()
```

```python
movie_budget = pd.read_csv('movies_budget_df.csv')
movie_budget.head()
```

```python
movie_budget_revenue = pd.read_csv('movies_budget_df_revenue.csv')
movie_budget_revenue.head()
```

```python
merge_revenue = pd.merge(movie_budget, movie_budget_revenue, on="id")
merge_revenue.head()
```

```python
movies_budget_revenue_df = pd.merge(merge_revenue, movie_num, on="title")
movies_budget_revenue_df.head()
```

## Step II:  Clean and Wrangle


###  **netflix_revenue_df**

- Inspect dataframe
- Check for missing values
- Drop Nan values and not zero values !IMPORTANT
- Convert numeric columns to numeric if necessary
- Save to a new dataframe and export as csv

###  **movies_budget_revenue_df**

- Inspect dataframe
- Check for missing values
- Drop Nan values and not zero values !IMPORTANT
- Remove $ and (,) in numeric columns
- Convert numeric columns to numeric if necessary
- Save to a new dataframe and export as csv

```python
netflix_revenue_df.dropna()
netflix_revenue_df = netflix_revenue_df.loc[netflix_revenue_df["revenue"]!= 0,:]
netflix_revenue_index_df = netflix_revenue_df.reset_index()
new_netflix_revenue_df = netflix_revenue_index_df[['title','revenue','popularity']]
new_netflix_revenue_df
```

```python
movies_budget_revenue_df.dropna()
movies_budget_revenue_df = movies_budget_revenue_df.loc[movies_budget_revenue_df["revenue"]!= 0,:]
movies_budget_revenue_index_df = movies_budget_revenue_df.reset_index()
new_movies_budget_revenue_df = movies_budget_revenue_index_df[['title','revenue','ProductionBudget']]
new_movies_budget_revenue_df
```
